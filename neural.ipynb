```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([3,6,9,11,8,4])
y = np.array([1,2,3,4,5,6])
plt.scatter(x,y)
plt.show()
```

```python
model=Sequential()
model.add(Dense(1,input_dim=1,activation='linear'))
model.add(Dense(1,activation='linear'))
model.compile(Adam(learning_rate=0.01),loss='mse')
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=20,verbose=1)
```

```python
model.predict(x)
```

```python
import pandas as pd
import seaborn as sns
```

```python
iris = pd.read_csv(r'/content/Iris.csv')
iris
```

```python
iris = iris.set_index('Id')
iris
```

```python
iris = iris.rename(columns={'SepalLengthCm':'SL','SepalWidthCm':'SW','PetalLengthCm':'PL','PetalWidthCm':'PW','Species':'Flower'})
iris
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
iris.Flower = le.fit_transform(iris.Flower)
```

```python
iris
```

```python
sns.displot(iris.SL[iris.Flower==0])
sns.displot(iris.SL[iris.Flower==1])
sns.displot(iris.SL[iris.Flower==2])
plt.show()
```

```python
sns.distplot(iris.SL[iris.Flower==0])
sns.distplot(iris.SL[iris.Flower==1])
sns.distplot(iris.SL[iris.Flower==2])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from keras.utils import to_categorical
```

```python
ip = iris.drop('Flower',axis=1)
op = iris.Flower
```

```python
op = to_categorical(op,3)
```

```python
x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(50,input_dim=4,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.compile(Adam(learning_rate=0.01),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
h = model.fit(x_train,y_train,epochs=20,validation_data=(x_test,y_test))
```

```python
yp = model.predict(x_test)
```
